# Power BI Sales Project #

### Here's a brief overview of how I started with 4 csv files of quarterly sales, and ended up with this report:

## ETL

The first part was loading the datasets into Power BI. After they were loaded in, I checked the data for errors,using the data integrity tool available in the view ribbon. I got rid of error/null values and made sure all the data types were correct for the columns. There weren't too many items to fix at this stage.


Then, because each table was for a different sales quarter from the same year-- and the data was all in identical columns with same names and data types-- I appended the tables.

After the queries were appended, I decided I didn't need all the information of the order addresses (for my visualizations) what I really needed for that column-- was the city the order was placed in. So using extract, I extracted the city and changed the column name to fit the values it now held.


Next, because of Power BI's set-up in the report view (and for good practice) I configured the appended query into a schema. There weren't too many columns-- but I was able to make a fact table of the order details, as well as 2 dim tables: for products and orders. After the schema was completed it was time to close the query editor, and apply the changes I made.

# Report view

Once I was in the report view, I went to the data model, to connect the different tables so that I would be able to make my visualizations.

I then went into the data view to make a few calculations that I could use for visualizing the data. Using DAX, I made a calculated column for revenue, as well as some measures that might come in handy later. I also changed the data type of the Order City from uncategorized to city-- which would help if I decidedto use any geospatial charts in the report.

# Visualizations

Last, I created the report. I wanted to use different kinds of charts to really show what Power BI could do. I first made a line chart of the total sales by quarter, and column chart of the products (sorted in descending order).

Next, I wanted to use the fancy map to show sales by city. I also used a card to put the total sales, used a multiline card for the top 5 sales by city, and put a pie chart of the top 5 grossing products.

In the next page, I added a table of a unique list of producs, as well as their price. I put that next to an area chart showing the volume of product sold-- and added a slicer so that the user could check out the different volume of sales in different months. Because of the hierarchy of the order date, there were a lot of different ways I could've displayed this information-- but the one I went with looked the most interesting for this report. 

On the final page, I added a couple more cards of some measures I'd made with DAX, as well as a stacked bar of product sales by weekday. Because the order dates included the day-- I had gone back into the data view to create another formula using the format( ) function-- to get the days of the week, and then using the weekday( ) function, created a column where those days were assigned ordered numerical values-- so that they could be sorted in the view, Monday- Sunday.

# Questions

I didn't have many questions going into this project, as it was pretty straightforward sales data by quarter. I did wonder which city had ordered the most, which product was the biggest seller, and grossed the most-- all those questions were answered. Surprisingly, San Francisco ordered the most from this retailer-- it was as much as the next 2 (LA and NYC put together). Also the biggest grosser was the Macbook followed by the iPhone (not too much of a surprise as they were the most expensive)-- althought the item that _sold the most_ were AAA batteries (which were also the lowest priced item).
